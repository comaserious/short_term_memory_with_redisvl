```python
schema = {
    "index" : {
        "name" : "chatbot-memory", # 인덱스 이름
        "prefix" : "conversation", # Redis에 저장될 키의 접두사
        "storage_type" : "json" # JSON storage type default 는 hash
    },
    "fields" : [
        # 1. thread_id: 대화를 구분하는 ID. 필터링을 위해 'tag' 타입 사용
        {"name" : "thread_id", "type" : "tag"},

        # 2. role: 메시지 발화자 ('user' 또는 'bot'). 필터링에 사용 가능
        {"name" : "role", "type" : "tag"},

        # 3. content: 실제 대화 내용 
        {"name" : "content", "type" : "text"},

        # 4. timestamp: 대화 순서 정렬을 위한 시간
        {"name" : "timestamp", "type" : "numeric"},

        # 5. embedding : content 를 벡터로 변환한 값 (유사도 검사용)
        {
            "name" : "embedding",
            "type" : "vector",
            "attrs" : {
                "dims" : 1536,
                "distance_metric": "cosine",
                "algorithm": "FLAT",
                "datatype": "FLOAT32"
            }
        }
    ]
}
```

ConversationMemory

```python
import time
import numpy as np
from redisvl.index import SearchIndex
from redisvl.query import VectorQuery
from redisvl.query.filter import Tag
from langchain_openai import OpenAIEmbeddings

class ConversationMemory:
    def __init__(self, thread_id : str, redis_url : str = "redis://localhost:30000"):
        self.thread_id = thread_id
        self.index = SearchIndex.from_dict(schema, redis_url = redis_url)
        # self.index.connect(redis_url)

        self.embedding_model = OpenAIEmbeddings()

        if not self.index.exists():
            print("인덱스가 존재하지 않아 새로 생성합니다.")

            self.index.create(overwrite=True)

    def _get_embedding(self, text : str):
        # 1. LangChain 모델로 임베딩 실행 -> float 리스트 반환
        vector_list = self.embedding_model.embed_query(text)

        # 2. 리스트를 float32 타입의 numpy 배열로 변환 (필수!)
        vector_np = np.array(vector_list, dtype=np.float32)

        # 3. numpy 배열을 bytes로 변환하여 반환
        return vector_np.tobytes()
    
    def add_message(self, role : str, content : str):
        print(f"[{self.thread_id}] 대화 추가: ({role}) {content}")

        message = {
            "thread_id" : self.thread_id,
            "role" : role,
            "content" : content,
            "timestamp": time.time(),
            "embedding" : self._get_embedding(content)
        }

        self.index.load([message])

    def search_similar_messages(self, query : str, top_k : int = 3):
        print(f"[{self.thread_id}] '{query}'와 유사한 대화 검색...")
        query_embedding = self._get_embedding(query)
        filter_expression = (Tag("thread_id") == self.thread_id)
        query = VectorQuery(
            vector = query_embedding,
            vector_field_name = "embedding",
            return_fields = ["role", "thread_id", "content", "timestamp", "vector_distance"],
            num_results = top_k,
            filter_expression = filter_expression
        )

        results = self.index.query(query)

        return results
    
    def delete(self):
        self.index.delete()
```

```python
print("--- 대화 시작: thread-A ---")
thread_a_memory = ConversationMemory("thread-A")
```

```python
print("--- 대화 시작: thread-A ---")

thread_a_memory.add_message("user", "오늘 날씨 어때?")
thread_a_memory.add_message("bot", "서울은 오늘 맑고 화창합니다.")
thread_a_memory.add_message("user", "그러면 제주도 날씨는?")
thread_a_memory.add_message("bot", "제주도는 현재 비가 오고 있습니다.")
thread_a_memory.add_message("user", "내이름은 고길동이야 기억해둬")
thread_a_memory.add_message("bot", "안녕하세요 고길동님 기억해두겠습니다.")

print("-" * 20)

# --- 두 번째 대화 (thread-B) ---
print("--- 대화 시작: thread-B ---")
thread_b_memory = ConversationMemory("thread-B")

# 대화 내용 저장
thread_b_memory.add_message("user", "점심 메뉴 추천해줘.")
thread_b_memory.add_message("bot", "오늘은 김치찌개 어떠신가요?")
thread_b_memory.add_message("user", "내 이름은 홍길동이야 기억해둬")
thread_b_memory.add_message("bot", "안녕하세요 홍길동님 기억해두겠습니다.")
```

```python
results_a = thread_a_memory.search_similar_messages("내 이름이 뭐라고??")
results_b = thread_b_memory.search_similar_messages("내 이름이 뭐라고?")

for doc in results_a:
    # print(f"[{doc.thread_id}] {doc.role}: {doc.content} (유사도: {doc.vector_distance})")
    print(doc)

print("-" * 20)

for doc in results_b:
    # print(f"[{doc.thread_id}] {doc.role}: {doc.content} (유사도: {doc.vector_distance})")
    print(doc)

print("-" * 20)

# --- 세 번째 대화 (thread-C) ---
```
